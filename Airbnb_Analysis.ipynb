```python
import pandas as pd

# ✅ Correct file path (yours)
DATA_PATH = r"C:\Users\utsha\Desktop\AIRBNB HOTEL BOOKING ANALYSIS\Airbnb Dataset.xlsx"

# Load the Excel file
xls = pd.ExcelFile(DATA_PATH)

# ✅ Use print() to show sheet names
print("Sheets in Excel file:", xls.sheet_names)

# Load the first sheet (change index if needed)
df = pd.read_excel(xls, xls.sheet_names[0])

# Display top 5 rows
df.head()
```

```python
# Check structure
df.info()

# Check basic statistics
df.describe()

# Check missing values
df.isnull().sum()
```

```python
# Step 2: Data Cleaning and Basic Exploration

# 📌 1. View basic info about the dataset
print("----- Dataset Info -----")
df.info()

# 📌 2. Check the first few rows
print("\n----- First 5 Rows -----")
display(df.head())

# 📌 3. Check for missing values
print("\n----- Missing Values (per column) -----")
print(df.isnull().sum())

# 📌 4. Check number of duplicate rows
duplicates = df.duplicated().sum()
print(f"\n----- Total Duplicate Rows: {duplicates} -----")

# 📌 5. Display column names
print("\n----- Column Names -----")
print(list(df.columns))

# 📌 6. Basic statistics for numeric columns
print("\n----- Summary Statistics -----")
display(df.describe())

# 📌 7. Optional: Remove duplicates (if any)
df.drop_duplicates(inplace=True)

# 📌 8. Optional: Clean up column names (remove spaces, lowercase)
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

print("\n✅ Data cleaning step completed successfully!")
print("Updated columns:", list(df.columns))
```

```python
# Step 3: Data Understanding & Visual Analysis

import matplotlib.pyplot as plt
import seaborn as sns

# ✅ Set plot style
plt.style.use('seaborn-v0_8')
sns.set_palette("viridis")

# -------------------------------
# 1️⃣ Property Type Distribution
# -------------------------------
if 'room_type' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x='room_type', order=df['room_type'].value_counts().index)
    plt.title("Distribution of Property Types")
    plt.xlabel("Room Type")
    plt.ylabel("Count")
    plt.xticks(rotation=20)
    plt.show()
elif 'property_type' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x='property_type', order=df['property_type'].value_counts().index[:10])
    plt.title("Top 10 Property Types")
    plt.xlabel("Property Type")
    plt.ylabel("Count")
    plt.xticks(rotation=20)
    plt.show()

# -------------------------------
# 2️⃣ Neighborhood Group by Listing Count
# -------------------------------
if 'neighbourhood_group' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x='neighbourhood_group', order=df['neighbourhood_group'].value_counts().index)
    plt.title("Listings by Neighbourhood Group")
    plt.xlabel("Neighbourhood Group")
    plt.ylabel("Count")
    plt.xticks(rotation=25)
    plt.show()
elif 'neighbourhood' in df.columns:
    top_neigh = df['neighbourhood'].value_counts().head(10)
    plt.figure(figsize=(8, 5))
    sns.barplot(x=top_neigh.values, y=top_neigh.index)
    plt.title("Top 10 Neighbourhoods by Number of Listings")
    plt.xlabel("Number of Listings")
    plt.ylabel("Neighbourhood")
    plt.show()

# -------------------------------
# 3️⃣ Average Price by Neighbourhood Group
# -------------------------------
if 'price' in df.columns:
    if 'neighbourhood_group' in df.columns:
        plt.figure(figsize=(8, 5))
        sns.barplot(data=df, x='neighbourhood_group', y='price', estimator='mean')
        plt.title("Average Price by Neighbourhood Group")
        plt.xlabel("Neighbourhood Group")
        plt.ylabel("Average Price")
        plt.xticks(rotation=25)
        plt.show()
```

```python
# Check if columns exist
if 'year_built' in df.columns and 'price' in df.columns:
    import matplotlib.pyplot as plt
    import seaborn as sns
    
    plt.figure(figsize=(8,5))
    sns.scatterplot(x='year_built', y='price', data=df)
    plt.title("Relationship Between Construction Year and Price")
    plt.xlabel("Year Built")
    plt.ylabel("Price")
    plt.show()

    # Optional: Check correlation
    correlation = df['year_built'].corr(df['price'])
    print("Correlation between Year Built and Price:", correlation)
```

```python
# Check if columns exist
if 'host_id' in df.columns and 'calculated_host_listings_count' in df.columns:
    
    # Group by host_id and sum listings count
    top_hosts = df.groupby('host_id')['calculated_host_listings_count'].sum().sort_values(ascending=False).head(10)
    
    # Display top 10 hosts
    print("\nTop 10 Hosts by Total Listings:\n", top_hosts)
    
    # Plot bar chart
    import matplotlib.pyplot as plt
    import seaborn as sns
    
    plt.figure(figsize=(8,5))
    sns.barplot(x=top_hosts.values, y=top_hosts.index)
    plt.title("Top 10 Hosts by Listings Count")
    plt.xlabel("Total Listings")
    plt.ylabel("Host ID")
    plt.show()
```

```python
# Check if columns exist
if 'review_scores_rating' in df.columns:
    import matplotlib.pyplot as plt
    import seaborn as sns

    if 'neighbourhood_group' in df.columns and 'room_type' in df.columns:
        # Group by neighbourhood group and room type, calculate mean review scores
        avg_review = df.groupby(['neighbourhood_group','room_type'])['review_scores_rating'].mean().unstack()
        print("\nAverage Review Scores by Neighborhood & Room Type:\n", avg_review)

        # Plot grouped bar chart
        avg_review.plot(kind='bar', figsize=(10,6))
        plt.title("Average Review Scores by Neighborhood Group & Room Type")
        plt.xlabel("Neighborhood Group")
        plt.ylabel("Average Review Score")
        plt.xticks(rotation=25)
        plt.show()
```

```python

ARRN = df.groupby(['neighbourhood_group', 'room_type'])['review_rate_number'].mean().to_frame()

# Rename the column for clarity
ARRN.rename(columns={'review_rate_number': 'average_review_rate'}, inplace=True)

# Display the result
print("Average Review Rate Number (ARRN) by Neighborhood Group and Room Type:\n")
display(ARRN)
```
